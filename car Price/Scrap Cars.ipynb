```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException, SessionNotCreatedException

import time
import warnings
import pandas as pd
warnings.filterwarnings("ignore")
```

### Using Selenium

```python
driver=webdriver.Edge(r"C:\Users\jitsa\Downloads\edgedriver(2)\msedgedriver.exe")
driver.get("https://www.cars24.com")
```

```python
def search_car24():
    user=driver.find_element(By.XPATH,'//*[@id="react-select-2-input"]')
    u_search=str(input("Enter the search With location "))
    user.send_keys(u_search)
    
    #Submit Query
    search=driver.find_element(By.XPATH,'//span[@class="_1Aa4N"]')
    search.click()
    
search_car24()   
```

```python
# scrolling page to max

for i in range(20):
    driver.execute_script("window.scrollBy(0,2000)")
    
    time.sleep(2)
```

```python
url=[]

for i in driver.find_elements(By.XPATH,'//div[@class="col-4"]/a'):
    url.append(i.get_attribute('href'))
    
len(url)
```

# Parameters

- Brand, model, variant, manufacturing year, driven kilometers, fuel, number of owners, location and last target variable Price

```python
Brand=[]
Model=[]
Varient=[]
Make_year=[]
K_meter=[]
F_type=[]
Owner=[]
Location=[]
Price=[]


for url in url:
    driver.get(url)
    time.sleep(2)
    
# Brand name
    try:
        brand = driver.find_element(By.XPATH,'//li[@class="breadcrumb-item"]/a')
        Brand.append(brand.text.strip())
    except NoSuchElementException:
        Brand.append("-")
        
# Model
    try:
        m=driver.find_element(By.XPATH,'//h2[@class="_2geSF"]')
        Model.append((m.text.split()[2]))
    except NoSuchElementException:
        Model.append('-')
        
# Varient 
    try:
        var=driver.find_element(By.XPATH,'//span[@class="_18vo2"]')
        Varient.append(var.text.strip())
    except NoSuchElementException:
        Varient.append('-')
        
# Make year
    try:
        mk=driver.find_element(By.XPATH,'//ul[@class="_3oLON"]/li[5]')
        Make_year.append(mk.text.split()[0])
    except NoSuchElementException:
        Make_year.append('-')
        
#-------------------------------------------------------------------------------------------------------------------------------
    
# Fuel Type
    try:
        ft= driver.find_element(By.XPATH,'//ul[@class="_2tp8h"]/li[3]')
        F_type.append(ft.text.strip())
    except NoSuchElementException:
        F_type.append("-")
        
# Kilo Meter
    try:
        km=driver.find_element(By.XPATH,'//ul[@class="_2tp8h"]/li[1]')
        K_meter.append(km.text.strip())
    except NoSuchElementException:
        k_meter.append('-')
        
# Owner
    try:
        o=driver.find_element(By.XPATH,'//ul[@class="_2tp8h"]/li[5]')
        Owner.append(o.text.strip())
    except NoSuchElementException:
        Owner.append('-')
        
# Location
    try:
        l=driver.find_element(By.XPATH,'//ul[@class="_2tp8h"]/li[7]')
        Location.append(l.text.strip())
    except NoSuchElementException:
        Location.append('-')
        
# Price
    try:
        p=driver.find_element(By.XPATH,'//ul[@class="_10Etm"]/li[1]/span[2]')
        Price.append(p.text.strip("â‚¹"))
    except NoSuchElementException:
        Price.append('-')
        
    
    time.sleep(1)
```

```python
print("Brand",len(Brand),"Model",len(Model),"Varient",len(Varient),"Year",len(Make_year),"Kilometer",len(K_meter),"Fuel",len(F_type),"Owner",len(Owner),"Location",len(Location),"Price",len(Price))
```

```python
df=pd.DataFrame(data={"Brand":Brand,"Model":Model,
                      "Varient":Varient,"Make_Year":Make_year,
                      "Kilometer":K_meter,"Fuel_Type":F_type,
                      "Owners":Owner, "Location":Location,"Price":Price})
```

```python
df.head()
```

```python
df.to_csv('delhi.csv', sep =',')
```

```python
driver.close()
```
