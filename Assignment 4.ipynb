1. Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:

A. Rank
B. Name
C. Artist
D. Upload date
E. Views


```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
from selenium.webdriver.support.wait import WebDriverWait
import time
import pandas as pd
```

```python
driver=webdriver.Edge(r'C:\Users\jitsa\Downloads\msedgedriver.exe')
driver.get(r"https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
Rank=[]

for i in driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[1]'):
    Rank.append(i.text.strip(""))
print(len(Rank),Rank)
```

```python
Name=[]
for i in driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[2]'):
    Name.append((i.text.strip("]")).replace('[',' '))
    
print(len(Name),Name)
```

```python
Artist=[]
for i in driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[3]'):
    Artist.append(i.text.strip(" "))
print(len(Artist),"\n",Artist)
```

```python
date=[]
for i in driver.find_elements(By.XPATH,'//td[@align="right"]')[0:30]:
    date.append(i.text.strip())
print(len(date),date)
```

```python
Views=[]

for i in driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[4]'):
    Views.append(i.text.strip("\n"))
    
print(len(Views),Views)
```

```python
Notes=[]

for i in driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[6]'):
    Notes.append(i.text.strip(" "))

print(len(Notes),Notes)
```

```python
df1=pd.DataFrame({'Rank':Rank,'Video Name':Name,'Artist':Artist,'Publish Date':date,'Views':Views,'Notes':Notes})
```

# Q7

Scrape the details of Data science recruiters from naukri.com. Url = https://www.naukri.com/. You have to find the following details:

- Name
- Designation
- Company
- Skills they hire for
- Location

Note: - From naukri.com homepage click on the recruiters option and the on the search pane type Data science and click on search. All this should be done through code

```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
from selenium.webdriver.support.wait import WebDriverWait
import time
import pandas as pd
```

```python
driver=webdriver.Edge(r'msedgedriver.exe')
driver.get(r' https://www.naukri.com/')
```

```python
skill=driver.find_element(By.XPATH,'//ul[@class="nI-gNb-menus"]/li[1]')
skill.click()
```

```python
input1=driver.find_element(By.XPATH,'//*[@id="root"]/div[6]/div/div/div[1]/div/div/div/input')
input1.send_keys("Data Science")
```

```python
search=driver.find_element(By.XPATH,'//*[@id="root"]/div[6]/div/div/div[6]')
search.click()
```

```python
Title_url=[]
for i in driver.find_elements(By.XPATH,'//a[@class="title fw500 ellipsis"]'):
    Title_url.append(i.get_attribute('href'))
len(Title_url)
```

```python
Job=[]
Company=[]
Skill=[]
Location=[]

start=0
end=10

for i in range(start,end):
    for i in driver.find_elements(By.XPATH,'//a[@class="title fw500 ellipsis"]'):
        Job.append(i.text.strip())
    for i in driver.find_elements(By.XPATH,'//div[@class="info fleft"]/div/a[1]'):
        Company.append(i.text.strip())
    for i in driver.find_elements(By.XPATH,'//ul[@class="tags has-description"]'):
        Skill.append((i.text.strip(" ")).replace("\n"," ,"))
    for i in driver.find_elements(By.XPATH,'//ul[@class="mt-7"]/li[3]/span'):
        Location.append(i.text.strip())
   
    next_button=driver.find_element(By.XPATH,'//div[@class="fleft pages"]/a') #path of next button
    next_button.click() # click
    
    time.sleep(3)
```

```python
print(len(Job),len(Company),len(Skill),len(Location))
```

```python
print(Job,"\n",Company,"\n",Skill,"\n",Location)
```

```python
df7=pd.DataFrame(data={'Company_Name':Company,'Job':Job,'Location':Location,'Skill':Skill})
```

```python
df7
```

```python
driver.close()
```
